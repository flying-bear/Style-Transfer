# Datasets

```python
class RandomCropsDataset(Dataset):
  def __init__(self, path, width=512, crop=(224, 224)):
    self.files = [str(file) for file in pathlib.Path(path).iterdir()]
    if width is not None:
      self.transform = T.Compose([
        T.ToTensor(),
        T.Resize(width),
        T.RandomCrop(crop),
        T.ConvertImageDtype(torch.float),
      ])
    else:
      self.transform = T.Compose([
        T.ToTensor(),
        # T.Resize(width),
        T.RandomCrop(crop),
        T.ConvertImageDtype(torch.float),
      ])

  def __len__(self):
    return len(self.files)

  def __getitem__(self, idx):
    try:
      image = cv2.imread(self.files[idx])
      image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
      return self.transform(image), 0
    except:
      print(f'failed to read {self.files[idx]}')
      image = cv2.imread(self.file[0])
      image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
      return self.transform(image), 0


class ContentStyleDataset(Dataset):
  """Return a random pair of (content, style)""" 
  def __init__(self, content_dataset, style_dataset, pairs_per_epoch=1000):
    self.content = content_dataset 
    self.style = style_dataset 
    self.size = pairs_per_epoch 

  def __len__(self):
    return self.size 

  def __getitem__(self, idx):
    i = np.random.randint(0, len(self.content))
    j = np.random.randint(0, len(self.style))
    content, _ = self.content[i]
    style, _ = self.style[j]
    return content, style
```

## Model

### Encoder

```python
from torchvision.models import vgg19 as vgg

class Encoder(nn.Module):
  def __init__(self):
    super().__init__()

    model = vgg(pretrained=True)

    self.preprocess = T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    self.encoder_0 = nn.Sequential(*model.features[:2])     # relu1_1
    self.encoder_1 = nn.Sequential(*model.features[2:7])    # relu2_1
    self.encoder_2 = nn.Sequential(*model.features[7:12])   # relu3_1
    self.encoder_3 = nn.Sequential(*model.features[12:21])  # relu4_1

    for model in [getattr(self, f'encoder_{i}') for i in range(4)]:
      for param in model.parameters():
        param.requires_grad = False

  def forward(self, x):
    results = [self.preprocess(x)]
    for i in range(4):
      x = getattr(self, f'encoder_{i}')(results[-1])
      results.append(x)
    return results[1:]
```

```python
class UpScaleBlock(nn.Module):
  def __init__(self, in_dim, out_dim, n_conv_repeat):
    super().__init__()

    modules = [ 
        nn.Conv2d(in_dim, out_dim, (3, 3), padding=1, padding_mode='reflect'),
        nn.LeakyReLU(0.2),
        nn.Upsample(scale_factor=2, mode='nearest'),
    ]

    for _ in range(n_conv_repeat):
      modules += [
           nn.Conv2d(out_dim, out_dim, (3, 3), padding=1, padding_mode='reflect'), 
           nn.LeakyReLU(0.2)
      ]
    
    self.net = nn.Sequential(*modules)
  
  def forward(self, x):
    return self.net(x)
```

```python
class Decoder(nn.Module):
  def __init__(self):
    super().__init__()

    self.net = nn.Sequential(
        UpScaleBlock(512, 256, 3),
        UpScaleBlock(256, 128, 1),
        UpScaleBlock(128, 64, 1),
        nn.Conv2d(64, 3, (3, 3), padding=1, padding_mode='reflect'),
        nn.Sigmoid(),
    )

  def forward(self, x):
    return self.net(x)
```

```python
class Decoder(nn.Module):
  def __init__(self):
    super().__init__()

    self.net = nn.Sequential(
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(512, 256, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.Upsample(scale_factor=2, mode='nearest'),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(256, 256, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(256, 256, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(256, 256, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(256, 128, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.Upsample(scale_factor=2, mode='nearest'),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(128, 128, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(128, 64, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.Upsample(scale_factor=2, mode='nearest'),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(64, 64, (3, 3)),
        nn.LeakyReLU(0.2),
        nn.ReflectionPad2d((1, 1, 1, 1)),
        nn.Conv2d(64, 3, (3, 3)),
        nn.Sigmoid(),
    )

  def forward(self, x):
    return self.net(x)
```

### AdaIN Model

```python
def AdaIN(content_feats, style_feats):
  eps = torch.tensor(0.000001).float()
  content_mean = torch.mean(content_feats, (2, 3)).unsqueeze(2).unsqueeze(2).float() 
  content_std = torch.std(content_feats, (2, 3)).unsqueeze(2).unsqueeze(2).float()

  normed = (content_feats - content_mean) / (content_std + eps)  

  style_mean = torch.mean(style_feats, (2, 3)).unsqueeze(2).unsqueeze(2).float()
  style_std = torch.std(style_feats, (2, 3)).unsqueeze(2).unsqueeze(2).float()
  return normed * style_std + style_mean
```

```python
class AdaINModel(LightningModule):
  def __init__(self, saving_frequency=20):
    super().__init__()
    self.save_hyperparameters()

    self.encoder = Encoder()
    self.decoder = Decoder() 

    self.mse_loss = nn.MSELoss()
    self.bce_loss = nn.BCELoss()

  def forward(self, content, style, alpha=1.00):
    content_feats = self.encoder(content)[-1]
    style_feats = self.encoder(style)

    feats = content_feats
    feats = AdaIN(content_feats, style_feats[-1])
    a = torch.tensor(alpha).float()
    b = torch.tensor(1.0 - alpha).float()
    feats = a * feats + b * content_feats

    output = self.decoder(feats)
    return output, feats, style_feats 

  def style_loss(self, x, y):
    x_mean = torch.mean(x, (2, 3))
    x_std = torch.std(x, (2, 3))
    y_mean = torch.mean(y, (2, 3))
    y_std = torch.std(y, (2, 3))
    return self.mse_loss(x_mean, y_mean) + self.mse_loss(x_std, y_std)

  def training_step(self, batch, batch_idx):
    content, style = batch
    output, feats, style_feats = self(content, style)

    if self.global_step % self.hparams.saving_frequency == 0:
      torch.save(self.state_dict(), 'tmp.pth')
      grid = torchvision.utils.make_grid(
          torch.cat((style[:6], output[:6], content[:6]), 2)
          )
      dream_images = wandb.Image(grid)
      wandb.log({"dreams": dream_images}, step=self.global_step)


    output_feats = self.encoder(output)
    content_loss = self.mse_loss(output_feats[-1], feats)

    style_loss = 0
    for i in range(1, 4):
      style_loss += self.style_loss(output_feats[i], style_feats[i])

    wandb.log({
        'train/content-loss': content_loss.detach(),
        'train/style-loss': style_loss.detach(),
    }, step=self.global_step)

    return content_loss * 1000 + style_loss * 10000

  def training_epoch_end(self, outs):
    wandb.save('tmp.pth', policy='now')

  def configure_optimizers(self):
      return Adam(self.parameters(), lr=1e-4)
```

```python
def evaluate(model, path='test_data', output_path=None, scales=None, 
             copy_notebook=False,  copy_checkpoint=True, style_path='dataset/owenget-hr-prp'):
  if output_path is None:
    if IN_COLAB:
      output_path = '/content/drive/MyDrive/style-transfer/examples'
    else:
      output_path = 'examples'

  prefix = get_timecoded_name(model)
  output_path = pathlib.Path(output_path) / prefix
  output_path.mkdir(exist_ok=True, parents=True)

  if copy_checkpoint:
   save_checkpoint(model, path=output_path)
  
  if IN_COLAB:
    notebook_copy = str(output_path / 'style-transfer.ipynb')
    notebook_init = '/content/drive/MyDrive/style-transfer/style-transfer.ipynb'
    if copy_notebook:
      !cp $notebook_init $notebook_copy

  styles = list(pathlib.Path(style_path).iterdir())

  with torch.no_grad():
    model.cuda()
    for file in pathlib.Path(path).iterdir():
      try:
        image = read_image(str(file)).float() / 255.
        style = read_image(str(styles[np.random.randint(0, len(styles))])).float() / 255.
      except:
        print(f'fail to read file: {file}')
        continue 

      images = [] 
      stylized = []
      scales = sorted(scales)
      if scales is not None:
        for scale in scales:
          images.append(T.Resize(scale)(image))
        min_scale = images[0].shape[1:] 
      else:
        min_scale = image.shape[1:]
        images.append(image)

      for img in images:
        stylized.append(T.Resize(min_scale)(
            model(img.unsqueeze(0).cuda(), style.unsqueeze(0).cuda())[0][0].cpu())
        )

      name = str(uuid.uuid4())[:8] + '.png'
      fp = str(output_path / name)
      scaling_factor = min_scale[0] / style.shape[1]
      save_image(torch.cat((T.Resize(min_scale)(image), *stylized, T.Resize((min_scale[0], int(style.shape[2] * scaling_factor)))(style)), 2), fp)
      print(f'>> {fp} | {get_file_size_mb(output_path / name):.2f} MB')
```
